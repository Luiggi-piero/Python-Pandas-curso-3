```python
import pandas as pd
```

```python
data = [0.5, None, None, 0.52, 0.54, None, None, 0.59, 0.6, None, 0.7]
s = pd.Series(data)
s
```

```python
# rellenando los nulos con 0
s.fillna(0)
```

```python
s
```

```python
# ffill : inicia desde el principio, reemplaza el valor actual (si es nulo) por el valor anterior
s.fillna(method = 'ffill')
```

```python
# bfill: inicia desde el final, reemplaza el valor actual (si es nulo) por el valor anterior
s.fillna(method = 'bfill')
```

```python
s
```

```python
# reemplaza los nulos por la media de la serie
s.fillna(s.mean())
```

```python
# reemplaza el valor nulo solo del primer elemento que encuentre (inicia desde arriba/principio)
# si encuentra dos o más valores nulos seguidos, entonces solo reemplazará el primero, los siguientes no los toca
s.fillna(method = 'ffill', limit = 1)
```

```python
s
```

```python
# Combina los metodos ffill y bfill
# cambia los primeros nulos por el primer valor no nulo
# y
# cambia los ultimos nulos por el ultimo valor no nulo
s1 = s.fillna(method = 'ffill', limit = 1)
s1
```

```python
s2 = s1.fillna(method = 'bfill', limit = 1)
s2
```

```python
atletas = pd.DataFrame([
['Marcos', 9.62], ['Pedro', None], ['Juan', 9.69],
['Beto', 9.72], ['Sandro', None], ['Denis', 9.69],
['Ary', None], ['Carlos', 9.74]],
columns = ['Corredor', 'Mejor tiempo'])
atletas
```

```python
# reemplazando los valores nulos de la columna 'Mejor Tiempo' por la media de la misma columna
atletas.fillna(atletas['Mejor tiempo'].mean(), inplace = True)
```
