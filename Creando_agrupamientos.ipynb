```python
# elimina los duplicados y muestra los valores unicos
datos['Distrito'].drop_duplicates()
```

```python
grupo_barrio = datos.groupby('Distrito')
```

```python
type(grupo_barrio)
```

```python
grupo_barrio.groups
```

```python
  # media de cada grupo
  # data: es el dataframe que se genera por cada grupo
  # es decir, un dataframe para Ate otro para Barranco etc ...
for barrio, data in grupo_barrio:
    print('{} -> {}'.format(barrio, data.Valor.mean()))
```

```python
# lo mismo que lo anterior: media de cada grupo
grupo_barrio['Valor'].mean().round(2)
```

```python
# mediana de 2 columnas de cada grupo: media de la columna Valor y Mantenimiento
grupo_barrio[['Valor', 'Mantenimiento']].mean().round(2)
```

## Estadísticas descriptivas

```python
grupo_barrio['Valor'].describe()
```

```python
# solo obtener la estadistica de minimo y maximo, también se renombra
grupo_barrio['Valor'].aggregate(['min', 'max']).rename(columns = {'min': 'Mínimo', 'max': 'Máximo'})
```

```python
# usamos la biblioteca matplotlib para complementar a pandas
%matplotlib inline
import matplotlib.pyplot as plt

# creando un recurso(una figura)
plt.rc('figure', figsize = (20,10))
```

```python
# generando un grafico de barras para las medias de la columna Valor de los grupos de barrio
# objetivo: encontrar los puntos fuera de la curva (valores que no tienen sentido: por ejemplo comas es demasiado alto)
fig = grupo_barrio['Valor'].mean().plot.bar(color = 'blue')
fig.set_ylabel('Valor del alquiler')
fig.set_title('Valor medio del alquiler por distrito', {'fontsize': 22})
```

```python
# generando un grafico de barras para el valor máximo de la columna Valor de los grupos de barrio
# objetivo: encontrar los puntos fuera de la curva, es decir, valores que no tienen sentido: por ejemplo chorrillos es demasiado alto
fig = grupo_barrio['Valor'].max().plot.bar(color = 'blue')
fig.set_ylabel('Valor del alquiler')
fig.set_title('Valor máximo del alquiler por distrito', {'fontsize': 22})
```

## Ejemplo

```python
precios = pd.DataFrame([
['Feria', 'Cebolla', 2.5],
         ['Mercado', 'Cebolla', 1.99],
         ['Supermercado', 'Cebolla', 1.69],
         ['Feria', 'Tomate', 4],
         ['Mercado', 'Tomate', 3.29],
         ['Supermercado', 'Tomate', 2.99],
         ['Feria', 'Papa', 4.2],
         ['Mercado', 'Papa', 3.99],
         ['Supermercado', 'Papa', 3.69]],
columns = ['Local', 'Producto', 'Precio'])
precios
```

```python
productos = precios.groupby('Producto')
productos.describe().round(2)
```

```python
productos['Precio'].aggregate(['mean', 'std', 'min', 'max']).rename(columns = {'mean': 'Promedio', 'std': 'Desviación Estándar', 'min': 'Mínimo', 'max': 'Máximo'}).round(2)
```
