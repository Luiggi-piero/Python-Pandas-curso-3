## Inmuebles residenciales

```python
import pandas as pd
datos = pd.read_csv('alquiler.csv', sep = ';')
datos.head()
```

```python
list(datos['Tipo'].drop_duplicates())
```

```python
residencial = ['Habitación',
 'Casa',
 'Departamento',
 'Casa en condominio',
 'Casa comercial',
 'Casa de villa']
residencial
```

```python
# recorre el dataframe datos y revisa si el valor de la columna tipo se encuentra en la lista residencial
seleccion = datos['Tipo'].isin(residencial)
seleccion
```

```python
# datos_residencial = datos[criterio_de_seleccion]
datos_residencial = datos[seleccion]
datos_residencial
```

```python
list(datos_residencial['Tipo'].drop_duplicates())
```

```python
# cantidad de registros de datos_residencial
datos_residencial.shape[0]
```

```python
# cambiando los indices con numeros seguidos
datos_residencial.index = range(datos_residencial.shape[0])
datos_residencial
```
