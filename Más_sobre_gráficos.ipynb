```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
plt.rc('figure', figsize = (15, 10))
```

```python
datos = pd.read_csv('alquiler_residencial_sin_outliers.csv', sep = ';')
datos.head()
```

```python
# crear area
area = plt.figure()
```

```python
# add_subplot(num_filas, num_columnas, posicion_del_grafico)
# ubicación de los gráficos (g1, g2, g3 y g4)
g1 = area.add_subplot(2, 2, 1)
g2 = area.add_subplot(2, 2, 2)
g3 = area.add_subplot(2, 2, 3)
g4 = area.add_subplot(2, 2, 4)
```

```python
# Crear gráficos

# gráfico de dispersión
g1.scatter(datos.Valor, datos.Area)
g1.set_title('Valor X Área')

# gráfico histograma
g2.hist(datos.Valor)
g2.set_title('Histograma')

# gráfico de líneas
# seleccionando 100 lineas aleatorias(muestra)
datos_g3 = datos.Valor.sample(100)
# arreglando los indices para que sean secuenciales desde 0, 1, 2 ...
datos_g3.index = range(datos_g3.shape[0])

# g3.plt(datos_g3)
g3.plot(datos_g3)
g3.set_title('Muestra (Valor)')

# gráfico de barras
# agrupando por tipo y obteniendo la columna Valor
grupo = datos.groupby('Tipo')['Valor']
label = grupo.mean().index
valores = grupo.mean().values
g4.bar(label, valores)
g4.set_title('Valor medio por tipo')
```

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
plt.rc('figure', figsize = (15, 7))

datos = pd.read_csv('alquiler_muestra.csv', sep = ';')
```

```python
area = plt.figure()
g1 = area.add_subplot(1, 2, 1)
g2 = area.add_subplot(1, 2, 2)
grupo1 = datos.groupby('Tipo Agrupado')['Valor']
label = grupo1.count().index
valores = grupo1.count().values
g1.pie(valores, labels = label, autopct='%1.1f%%')
g1.set_title('Total de Inmuebles por Tipo Agrupado')
grupo2 = datos.groupby('Tipo')['Valor']
label = grupo2.count().index
valores = grupo2.count().values
g2.pie(valores, labels = label, autopct='%1.1f%%', explode = (.1, .1, .1, .1, .1))
g2.set_title('Total de Inmuebles por Tipo')
```
