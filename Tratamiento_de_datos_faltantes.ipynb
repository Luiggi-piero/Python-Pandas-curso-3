## Tratamiento de datos faltantes

```python
import pandas as pd
```

```python
datos = pd.read_csv('alquiler_residencial.csv', sep = ';')
```

```python
datos.head(10)
```

```python
datos.isnull()
```

```python
datos.notnull()
```

```python
datos.info()
```

```python
# muestra los registros con valores null de la columna Valor
datos[datos['Valor'].isnull()]
```

```python
A = datos.shape[0]
# elimina los valores nulos de la columna Valor
datos.dropna(subset = ['Valor'], inplace = True)
B = datos.shape[0]
A - B
```

```python
datos[datos['Valor'].isnull()]
```

```python
datos
```

```python
datos[datos['Mantenimiento'].isnull()].shape[0]
```

```python
seleccion = (datos['Mantenimiento'].isnull()) & (datos['Tipo'] == 'Departamento')
```

```python
A = datos.shape[0]
# ~ es la negacion
datos = datos[~seleccion]
B = datos.shape[0]
A - B
```

```python
datos[datos['Mantenimiento'].isnull()].shape[0]
```

```python
# rellena los elementos nulos con algún valor, lo busca en todas las columnas: datos.fillna(0)
# aqui lo busca solo en la columna Mantenimiento e Impuesto y lo reemplaza con 0
datos = datos.fillna({'Mantenimiento': 0, 'Impuesto': 0})
```

```python
# elementos nulos de la columna mantenimiento
datos[datos['Mantenimiento'].isnull()].shape[0]
```

```python
datos[datos['Impuesto'].isnull()].shape[0]
```

```python
datos.info()
```

```python
datos.to_csv('alquiler_residencial.csv', sep = ';', index = False)
```
