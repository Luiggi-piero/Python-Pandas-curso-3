## Creando variables

```python
import pandas as pd
```

```python
datos = pd.read_csv('alquiler_residencial.csv', sep = ';')
```

```python
datos.head(10)
```

```python
# nueva variable (nueva columna con la suma de las 3 columnas: Valor Mantenimiento Impuesto)
datos['Valor Bruto'] = (datos['Valor']) + (datos['Mantenimiento']) + (datos['Impuesto'])
datos.head(10)
```

```python
datos['Valor m2'] = (datos['Valor']) / (datos['Area'])
datos.head(10)
```

```python
# redondeando a 2 decimales
datos['Valor m2'] = datos['Valor m2'].round(2)
datos.head(10)
```

```python
datos['Valor Bruto m2'] = ((datos['Valor Bruto']) / (datos['Area'])).round(2)
datos.head(10)
```

```python
# los registros que sean de tipo Casa, Casa en condominio o Casa de villa ahora serán llamados Casa
casa = ['Casa', 'Casa en condominio', 'Casa de villa']
```

```python
# con apply aplicamos criterios en una columna
# x representará cada una de las filas de la columna Tipo
# si x esta en la lista casa entonces el valor será Casa, caso contrario será Departamento
datos['Tipo Agrupado'] = datos['Tipo'].apply(lambda x: 'Casa' if x in casa else 'Departamento')
datos.head(10)
```

```python
import pandas as pd
alumnos = pd.DataFrame({
'Nombre': ['Ary', 'Katia', 'Denis', 'Beto', 'Bruna', 'Dara', 'Carlos', 'Alice'],
'Sexo': ['M', 'F', 'M', 'M', 'F', 'F', 'M', 'F'],
'Edad': [15, 27, 56, 32, 42, 21, 19, 35],
'Notas': [7.5, 2.5, 5.0, 10, 8.2, 7, 6, 5.6]},
columns = ['Nombre', 'Edad', 'Sexo', 'Notas'])
```

```python
alumnos
```

```python
alumnos['Notas-Média(Notas)'] = alumnos['Notas'].apply(lambda x: x - alumnos['Notas'].mean())
alumnos
```

```python
alumnos['Rango Edad'] = alumnos['Edad'].apply(lambda x: 'Menor que 20 años' if x < 20
        else ('Entre 20 y 40 años' if (x >= 20 and x <= 40)
            else 'Mayor que 40 años'))
alumnos
```

```python
alumnos['Notas-Média(Notas)'] = alumnos.Notas - alumnos.Notas.mean()
alumnos
```
