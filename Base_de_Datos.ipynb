Concatenando dataframes

```python
data = [[1,2,3], [4,5,6], [7,8,9]]
data
```

```python
df1 = pd.DataFrame(data = data)
df1
```

```python
index = ['Linea' + str(i) for i in range(3)]
index
```

```python
df1 = pd.DataFrame(data = data, index = index)
df1
```

```python
columns = ['Columna' + str(i) for i in range(3)]
columns
```

```python
df1 = pd.DataFrame(data = data, index = index, columns = columns)
df1
```

```python
data = {'Columna0': {'Linea0': 1, 'Linea1': 4, 'Linea2': 7},
        'Columna1': {'Linea0': 2, 'Linea1': 5, 'Linea2': 8},
        'Columna2': {'Linea0': 3, 'Linea1': 6, 'Linea2': 9}}
data
```

```python
df2 = pd.DataFrame(data)
df2
```

```python
data = [(1,2,3), (4,5,6), (7,8,9)]
data
```

```python
df3 = pd.DataFrame(data = data, index = index, columns = columns)
df3
```

```python
# todos los valores del dataframe df1 que sean mayores a cero serán reemplazados por A
df1[df1 > 0] = 'A'
df1
```

```python
df2[df2 > 0] = 'b'
df2
```

```python
df3[df3 > 0] = 'C'
df3
```

```python
# usa las columnas como llave (el dataframe resultante crecerá verticalmente)
df4 = pd.concat([df1, df2, df3])
df4
```

```python
# usa las filas como llave (el dataframe resultante crecerá horizontalmente)
df5 = pd.concat([df1, df2, df3], axis = 1)
df5
```
